```python
personal_pronouns = ['I', 'i', 'mine', 'my', 'me', 'myself', 'Me', 'we', 'our']
```

```python
sample_texts = ["I don't understand why everything has to be so tough", 
               "I could kick someone right about now ugh, everything is sad and people are crazy", 
               "been a pretty good day so far but let's see, "]
```

```python
def predict_level(sample_text):
    
    print("*** RESULTS FOR TEXT: ", sample_text, " ***")
    positive_count=0;
    negative_count=0;
    personal_count=0;
    total_count = len(sample_text)

    str = sample_text.split()
    for word in str:
        if word in positive_words:
            positive_count += 1
        if word in negative_words:
            negative_count += 1
        if word in personal_pronouns:
            personal_count += 1

    minus_filler = positive_count + negative_count + personal_count
    positive_pc = 100*(positive_count/minus_filler)
    negative_pc = 100*(negative_count/minus_filler)
    personal_pc = 100*(personal_count/minus_filler)
    print("Positive: ", positive_pc, "%")
    print("Negative: " , negative_pc, "%")
    print("Personal pronouns: " , personal_pc, "%")
    
    self_score=0;
    if(personal_pc<=10):
        self_score = 0
    if(personal_pc>10 and personal_pc<=20):
        self_score = 1
    if(personal_pc>20 and personal_pc<=40):
        self_score = 2
    if(personal_pc>40 and personal_pc<=60):
        self_score = 3
    if(personal_pc>60 and personal_pc<=80):
        self_score = 4
    if(personal_pc>80):
        self_score = 5

    negative_score=0;
    if(negative_pc<=10):
        negative_score = 0
    if(negative_pc>10 and negative_pc<=20):
        negative_score = 1
    if(negative_pc>20 and negative_pc<=40):
        negative_score = 2
    if(negative_pc>40 and negative_pc<=60):
        negative_score = 3
    if(negative_pc>60 and negative_pc<=80):
        negative_score = 4
    if(negative_pc>80):
        negative_score = 5

    depressed_pc = (negative_score)*0.8 + (self_score)*0.2
    print(depressed_pc)
    
    if depressed_pc<=1:
        result = "Not Applicable"
    elif depressed_pc<=2:
        result = "Least"
    elif depressed_pc<=3:
        result = "Moderate"
    elif depressed_pc<=4:
        result = "High"
    elif depressed_pc<=5:
        result = "Severe"
    
    print("Diagnosis result: ", result)
    print()
```

```python
for s in sample_texts:
    predict_level(s)
```
